```python
from PIL import Image
import os

# Folder containing images
folder = "D:\ML\Self Driving Car\self_driving_car\Self-Driving-Car-Python\driving_data\gif"

# Get all image file paths (png/jpg/etc.), sorted alphabetically
image_files = sorted([
    os.path.join(folder, f)
    for f in os.listdir(folder)
    #if f.lower().endswith((".png", ".jpg", ".jpeg"))
])

# Open images and convert to RGBA
images = [Image.open(f).convert("RGBA") for f in image_files]

# Optional: resize all images to the size of the first image
width, height = images[0].size
images = [img.resize((width, height)) for img in images]

# Save as GIF
images[0].save(
    "output.gif",
    save_all=True,
    append_images=images[1:],
    duration=250,  # milliseconds per frame
    loop=0         # 0 = loop forever
)

print("GIF created successfully!")
```
